1. Can we use Bagging for regression problems?





Yes, Bagging can be used for regression. Instead of voting (as in classification), it averages the predictions of multiple regression models to give the final output.





2. Difference between multiple model training and single model training?




•Single model training: One model is trained on the dataset.


•Multiple model training: Many models are trained and their results are combined.
Multiple models usually give better accuracy and stability.




3. Feature randomness in Random Forest


Random Forest randomly selects a subset of features at each split.
This ensures trees are less similar, improving performance and reducing overfitting.




4. What is OOB (Out-of-Bag) Score



OOB score evaluates the model using data not used during training of each tree.



It works like internal validation, so no separate test set is needed.



5. How is feature importance measured in Random Forest?



By checking how much each feature reduces impurity (like Gini or MSE) across all trees.
Higher reduction more important feature.



6. Working principle of Bagging Classifier



•Create many bootstrap samples



•Train a model on each sample


•Combine predictions using majority voting




7. How to evaluate Bagging Classifier performance?




Using metrics like:

•Accuracy

•Precision

•Recall

•F1-score

•ROC-AUC



8. How does a Bagging Regressor work?




It trains multiple regression models on different samples and averages their outputs to reduce error.



9. Main advantage of ensemble techniques



They reduce overfitting and improve accuracy by combining multiple models.



10. Main challenge of ensemble methods



They are:



•Computationally expensive
•Harder to interpret compared to single models



11. Key idea behind ensemble techniques



Many weak learners together form a strong learner.



12. What is a Random Forest Classifier?



An ensemble of decision trees where:


Each tree is trained on random data
Final output is based on majority voting

14. What is ensemble learning?



A technique where multiple models are combined to solve a problem more accurately.



15. When should we avoid ensemble methods?



Very small datasets
When model interpretability is critical
Limited computational resources




16. How does Bagging reduce overfitting?



By training models on different subsets, errors cancel out and variance decreases.




17. Why is Random Forest better than a single Decision Tree?




Because it:


•Reduces overfitting.

•Handles noise better.

•Gives more stable predictions.



18. Role of bootstrap sampling in Bagging




It creates multiple datasets with replacement, increasing diversity among models.






19. Real-world applications of ensemble techniques



•Fraud detection

•Medical diagnosis

•Stock price prediction


•Recommendation systems





20. Difference between Bagging and Boosting.



Bagging-


•Models trained independently


•Reduces variances


•Equal weight to all samples


Boosting-


•Models trained sequentially


•Reduces bias


•More focus on misclassified samples
